### One Hot Encoding with many categories in a feature

```python
df2 = pd.read_csv('mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df2.head()
```

```python
for i in df2.columns:
    print(len(df2[i].unique()))
```

```python
df2.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
lst_10=df2.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)
```

```python
lst_10
```

```python
for categories in lst_10:
    df2[categories]=np.where(df2['X1']==categories,1,0)
```

```python
df2.head()
```

```python
df2[lst_10]
```

### ordinal number encoding

```python
import datetime
```

```python
today_date=datetime.datetime.today()
today_date
```

```python
today_date-datetime.timedelta(3)
```

```python
days=[today_date-datetime.timedelta(x) for x in range(0,15)]
```

```python
data=pd.DataFrame(days)
data.columns=["Day"]
```

```python
data.head()
```

```python
data['weekday']=data['Day'].dt.day_name()
data.head()
```

```python
dictionary={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
data['weekday_ordinal']=data['weekday'].map(dictionary)
```

```python
data
```

### count and frequency encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

### Target Guided Ordinal Encoding

```python
df3=pd.read_csv('train.csv',usecols=['Cabin','Survived'])
```

```python
df3.head()
```

```python
df3['Cabin'].fillna('Missing',inplace=True)
```

```python
df3['Cabin']=df3['Cabin'].astype(str).str[0]
```

```python
df3.head()
```

```python
df3.groupby(['Cabin'])['Survived'].mean()
```

```python
ordinal_labels=df3.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df3['Cabin_ordinal_labels']=df3['Cabin'].map(ordinal_labels2)
df3.head()
```

### Mean Encoding

```python
mean_ordinal=df3.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df3['mean_ordinal_encode']=df3['Cabin'].map(mean_ordinal)
df3.head()
```
